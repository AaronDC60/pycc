```python
from lasers import gaussian_laser
from lasers import sine_square_laser
import matplotlib.pyplot as plt
import numpy as np
```

```python
F_str = 0.01
omega = 2.874
sigma = 0.446994
center = 2.5
E = gaussian_laser(F_str, omega, sigma, center)
```

```python
points = int(1e3)
xmin, xmax = 0, 10
xlist = list(map(lambda x: float(xmax - xmin)*x/points, range(points+1)))
ylist = list(map(lambda y: E(y), xlist))
plt.plot(xlist, ylist)
plt.show()
```
